```python
import psycopg2

# PostgreSQL connection parameters
DB_HOST = 'localhost'
DB_PORT = '5432'
DB_NAME = 'YouTube_Data'
DB_USER = 'postgres'
DB_PASSWORD = 'Sheila'

# Connect to the database
connection = psycopg2.connect(
    host=DB_HOST,
    port=DB_PORT,
    database=DB_NAME,
    user=DB_USER,
    password=DB_PASSWORD
)

# Create a cursor
cur = connection.cursor()

try:
    # Define ALTER TABLE statements to add primary keys and foreign keys
    alter1_statements = [
    'ALTER TABLE "Cities_Totals" ADD PRIMARY KEY ("Date");',
    'ALTER TABLE "Cities_Chart" ADD CONSTRAINT fk_date1 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Content_Type_Chart" ADD CONSTRAINT fk_date2 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Device_type_Chart" ADD CONSTRAINT fk_date3 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "New_Return_Chart" ADD CONSTRAINT fk_date4 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Operating_Sys_Chart" ADD CONSTRAINT fk_date5 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Sharing_service_Chart" ADD CONSTRAINT fk_date6 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Subscription_source_Chart" ADD CONSTRAINT fk_date7 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Subscription_status_Chart" ADD CONSTRAINT fk_date8 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Traffic_source_Chart" ADD CONSTRAINT fk_date9 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");',
    'ALTER TABLE "Geography_Table" ADD CONSTRAINT fk_date10 FOREIGN KEY ("Date") REFERENCES "Cities_Totals"("Date");'
    ]

    # Execute ALTER TABLE statements
    for statement in alter1_statements:
        cur.execute(statement)

    # Commit the changes
    connection.commit()

except Exception as e:
    print("Error:", e)
    # Rollback the transaction in case of error
    connection.rollback()

finally:
    # Close the cursor and connection
    cur.close()
    connection.close()
```
