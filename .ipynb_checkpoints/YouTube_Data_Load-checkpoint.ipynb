```python
# List to store indices of non-equal DataFrames
non_equal_indices = []

# Compare DataFrames directly
for i in range(len(all_dfs)):
    for j in range(i + 1, len(all_dfs)):
        if not all_dfs[i].equals(all_dfs[j]):
            non_equal_indices.append((i, j))

# Print indices of non-equal DataFrames
if non_equal_indices:
    print("Non-equal DataFrames:")
    for idx_pair in non_equal_indices:
        print(f"DataFrame {idx_pair[0]} and DataFrame {idx_pair[1]} are not equal.")
else:
    print("All DataFrames are equal.")
```
